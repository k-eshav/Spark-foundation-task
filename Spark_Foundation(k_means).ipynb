```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Iris.csv')
x = df.iloc[:,1:5].values
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i,init = 'k-means++',n_init = 10,random_state = 42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('clustering')
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.xticks(range(1,11))
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3,init = 'k-means++',n_init = 10,random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,2],s=100,c='violet',label = 'cluster1')
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,2],s=100,c='indigo',label = 'cluster2')
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,2],s=100,c='blue',label = 'cluster3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,2],s=300,c='red',label='centroids')
plt.show()
```
